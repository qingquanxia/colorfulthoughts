## Periodic boundary conditions

```python
N = 1000
x = np.linspace(0,1,N)
y = np.zeros_like(x)
y[:300] = np.linspace(0,1e-3, 300)
y[300:] = np.linspace(1e-3,0, 700)

plt.xlim(0,1)
plt.ylim(-1.1,1.1)
plt.xlabel('position (m)')
plt.ylabel('displacement (mm)')
plt.plot(x,y*1e3, label='original')

Y = fft(y)
#plt.plot(t, Y)

def approx_y(n): 
    return sum([ Y[i]*np.exp(2*np.pi*1j*i*x)/N for i in range(-n,n+1)])
plt.plot(x, approx_y(1)*1e3, label='n=1')
plt.plot(x, approx_y(2)*1e3, label='n=2')
plt.plot(x, approx_y(3)*1e3, label='n=3')
plt.plot(x, approx_y(20)*1e3, label='n=20')
plt.plot(x, approx_y(50)*1e3, label='n=50')
plt.legend()
```

```python
plt.scatter(np.arange(-50,51), 10*np.log10(np.abs(np.concatenate((Y[-50:], Y[:51]))/Y[0])**2), color='orange')

plt.xlim(-51, 51)
plt.xlabel('Frequency component')
plt.ylabel('Power relative to DC component (dB)')
```

```python
fig, (ax1,ax2) = plt.subplots(1, 2, sharey=True)
fig.set_size_inches(10, 6)

components = { i: Y[i]*np.exp(2*np.pi*1j*i*x)/N for i in range(-50,51) }

f = 329.63

def animate_func(frame): 
    if frame < 0: 
        t = 0 * 0.100
    elif frame <= 100: 
        t = frame * 0.100
    else: 
        t = 100 * 0.100
    fig.suptitle(f"Periodic boundary conditions -- Time: {t:.2f} ms")
    ax1.cla()
    ax2.cla()
    ax1.set_xlim(0,1)
    ax1.set_ylim(-1.1,1.1)
    ax1.set_xlabel('$x/L$')
    ax1.set_ylabel('displacement (mm)')
    ax1.set_title('First 10 frequency components')
    ax2.set_xlabel('$x/L$')
    ax2.set_title('Sum of first 50 harmonics')

    ax1.plot(x, 1e3*components[0]*np.exp(1j* j*2*np.pi*f*t))
    for j in range(1,11): 
        ax1.plot(x, 1e3*components[j]*np.exp(1j* j*2*np.pi*f*t) + 1e3*components[-j]*np.exp(-1j* j*2*np.pi*f*t))
    ax2.plot(x, 1e3*sum([components[j]*np.exp(1j* j*2*np.pi*f*t) for j in range(-50,51)]))

frames = range(-10, 111)
ani = animation.FuncAnimation(fig, animate_func, frames, interval=100, blit=False)
#ani.save(filename="foo.gif", writer="pillow")
ani.save(filename="colorfulthoughts/content/assets/periodic-wave.mp4", writer="ffmpeg")
```

# Discrete time diagrams

```python
def blank_diagram(fig_width=10, fig_height=10,
                  bg_color="white", color="black", hide_axis=True):
    fig = plt.figure(figsize=(fig_width, fig_height))
    ax = fig.add_axes((0, 0, 1, 1))
    ax.set_xlim(0, fig_width)
    ax.set_ylim(0, fig_height)
    ax.set_facecolor(bg_color)

    if hide_axis: 
        ax.tick_params(bottom=False, top=False,
                       left=False, right=False)
        ax.tick_params(labelbottom=False, labeltop=False,
                       labelleft=False, labelright=False)
    
    ax.spines["top"].set_color(color)
    ax.spines["bottom"].set_color(color)
    ax.spines["left"].set_color(color)
    ax.spines["right"].set_color(color)

    return fig, ax

fig, ax = blank_diagram()
# For points, do plt.scatter(x,y,size, color)
# For lines, do plot.plot(x,y,linewidth, color)

theta = np.linspace(0,2*np.pi)


# Upper left: fourier series
ax.text(2,9.5, 'Fourier series', fontsize=16, ha='center')
ax.plot(np.cos(theta) + 2, np.sin(theta) + 8, color='teal')
plt.scatter(np.linspace(1,3,9), np.zeros(9) + 6, color='orange')
ax.text(2.4,9, "time period: $T$", color='teal')
ax.text(2.4,6.2, "freq spacing: $1/T$", color='orange')

# Bottom left: DFT
ax.text(2,4.2, 'Discrete Fourier Transform', fontsize=16, ha='center')
theta = np.linspace(0, 2*np.pi, 10)
plt.scatter(.7*np.cos(theta) + 2, .7*np.sin(theta) + 3, color='teal')
plt.scatter(.7*np.cos(theta) + 2, .7*np.sin(theta) + 1, color='orange')
ax.text(2.4,3.7, r"time spacing: $\Delta t$", color='teal')
ax.text(2.4,1.7, r"freq spacing: $1/(N\Delta t)$", color='orange')

# Upper right: continuous time.
ax.text(8,9.5, 'Fourier transform', fontsize=16, ha='center')
plt.arrow(8,8, -1,0, head_width=0.1, color='teal')
plt.arrow(8,8, 1,0, head_width=0.1, color='teal')
plt.arrow(8,7, -1,0, head_width=0.1, color='orange')
plt.arrow(8,7, 1,0, head_width=0.1, color='orange')

# Lower right: DTFS
ax.text(8,4.2, 'Discrete Nonperiodic Time', fontsize=16, ha='center')
plt.scatter(np.linspace(7,9,9), np.zeros(9) + 3, color='teal')
ax.text(8,3.2, r"time spacing: $\Delta t$", color='teal')
plt.text(8,1.5, '?', fontsize=36, ha='center')

fig.savefig("colorfulthoughts/content/assets/fourier-domains.svg")
```

# Dirac comb

```python
subcarriers = 10
reals = np.array([.4,-.4,2,.2,.5,-2,-1,1,.1,-.1])
imags = np.array([0,-1,-.2,-.4,.6,.1,2,-1,.4,-.6])
positions = np.array([0.49929258, 0.49894022, 0.49258435, 0.49048766, 0.49806091,
       0.4965936 , 0.49786156, 0.50403394, 0.49372238, 0.49319523])
#positions = np.random.uniform(0.49,0.51)
widths = np.array([0.004, 0.1,0.02,0.02,0.005]*2)
t = np.linspace(0,1, 10000, endpoint=False)
x = sum([(reals[i]+1j*imags[i])*np.exp(-((t-(positions[i]))/widths[i])**2) for i in range(subcarriers)])
x = fftshift(x)
plt.plot(x)
X = fft(x)
plt.plot(X*50)
plt.xlim(0,200)
plt.plot([0,10000], [0,0])
```

```python
fig, ax = blank_diagram(10,10)

# Two horizontals
ax.plot([0,10], [2,2], color='black', linewidth=.5)
ax.plot([0,10], [7,7], color='black', linewidth=.5)

# Two verticals indicating periodicity of frequency convolution.
ax.plot([8,8], [0,10], color='black', linewidth=.5)
ax.plot([9,9], [0,10], color='black', linewidth=.5)

# Upper FT sign
ax.arrow(6,7, .5,0, color='red', linewidth=1, head_width=0.2)
ax.text(6.2,7.2, 'Fourier transform', ha='center', color='red')

# Lower FT sign
ax.arrow(6,2, .5,0, color='red', linewidth=1, head_width=0.2)
ax.text(6.2,2.2, 'Fourier transform', ha='center', color='red')

# continuous time
ax.plot(np.linspace(0,5,5000)+0.5, fftshift(x)[10000//2-2500:10000//2+2500] + 7, color='teal')

# sampled time
ax.vlines(np.linspace(0,5,200)+0.5, np.zeros(200)+2, fftshift(x)[::25][400//2-100:400//2+100]+2, color='teal')

# single freq
tmp = fftshift(X)[10000//2-200:10000//2+200]/1000
ax.plot(np.linspace(-1,2,400*3)+8, np.concatenate((np.zeros(400),tmp,np.zeros(400))) + 7, color='orange')

# convoluted freq
tmp = fftshift(X)[10000//2-200:10000//2+200]/1000
ax.plot(np.linspace(-1,2,400*3)+8, np.concatenate((tmp,tmp,tmp)) + 2, color='orange')

ax.text(3.5,3, 'spacing: $\Delta t$')

# Width of the periodization in frequency
ax.arrow(8.5,8.5, .5,0, head_width=0.1, length_includes_head=True)
ax.arrow(8.5,8.5, -.5,0, head_width=0.1, length_includes_head=True)
ax.text(8.5,8.7, '$1/\Delta t$')

ax.text(3,9.5, 'Time Domain', fontsize=16, ha='center')
ax.text(8.5,9.5, 'Freq Domain', fontsize=16, ha='center')

ax.arrow(3,5.5, 0,-1, head_width=0.2, linewidth=1, color='red')
ax.text(3.1,5, 'Multiply by $d_c(t/\Delta t)$', color='red')

ax.arrow(8.5,5.5, 0,-1, head_width=0.2, linewidth=1, color='red')
ax.text(8.6,5, 'Convolve by $d_c(f\Delta t)$', color='red')
```
